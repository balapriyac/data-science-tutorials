

## 1. Choose the Right Chart for Your Data



```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
data = sns.load_dataset("tips")

# Bar plot for categorical data
plt.figure(dpi=150)
sns.barplot(x="day", y="total_bill", data=data)
plt.title("Average Total Bill by Day")
plt.show()
```

```python
# Histogram for distribution
plt.figure(dpi=150)
sns.histplot(data["total_bill"], kde=True)
plt.title("Distribution of Total Bill Amounts")
plt.xlabel("Total Bill")
plt.ylabel("Frequency")
plt.show()
```

```python
# Scatter plot for correlation
plt.figure(dpi=150)
sns.scatterplot(x="total_bill", y="tip", data=data)
plt.title("Total Bill vs Tip Amount")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()
```

## 2. Use Color to Enhance Data Comprehension

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
data = sns.load_dataset("tips")

# Scatter plot with color differentiation by day
plt.figure(dpi=150)
sns.scatterplot(x="total_bill", y="tip", hue="day", data=data, palette="Set1")
plt.title("Total Bill vs Tip Amount by Day")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()
```

```python
# Custom color palette for a histogram
custom_palette = sns.color_palette("coolwarm", as_cmap=True)
plt.figure(dpi=150)
sns.histplot(data["total_bill"], kde=True, color=custom_palette(0.7))
plt.title("Distribution of Total Bill Amounts")
plt.xlabel("Total Bill")
plt.ylabel("Frequency")
plt.show()
```

## 3. Visualize Statistical Summaries


```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
data = sns.load_dataset("tips")

# Box plot for statistical summaries
plt.figure(dpi=150)
sns.boxplot(x="day", y="total_bill", data=data)
plt.title("Total Bill Distribution by Day")
plt.xlabel("Day")
plt.ylabel("Total Bill")
plt.show()
```

```python
# Violin plot for distribution and summaries
plt.figure(dpi=150)
sns.violinplot(x="day", y="total_bill", data=data)
plt.title("Total Bill Distribution by Day (Violin Plot)")
plt.xlabel("Day")
plt.ylabel("Total Bill")
plt.show()
```

## 4. Add Context with Annotations

```python
# Annotating the highest and lowest tips
max_tip = data.loc[data["tip"].idxmax()]
min_tip = data.loc[data["tip"].idxmin()]

plt.figure(figsize=(8, 6),dpi=150)
sns.scatterplot(x="total_bill", y="tip", data=data)

# Annotate max tip
plt.annotate(f"Max Tip: {max_tip['tip']}",
             xy=(max_tip["total_bill"], max_tip["tip"]),
             xytext=(max_tip["total_bill"] + 5, max_tip["tip"] + 2),
             arrowprops=dict(facecolor='green', arrowstyle="->"))

# Annotate min tip
plt.annotate(f"Min Tip: {min_tip['tip']}",
             xy=(min_tip["total_bill"], min_tip["tip"]),
             xytext=(min_tip["total_bill"] - 10, min_tip["tip"] - 2),
             arrowprops=dict(facecolor='red', arrowstyle="->"))

plt.title("Total Bill vs Tip Amount with Max/Min Annotations")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()
```

## 5. Combine Multiple Plots for Comprehensive Analysis

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
data = sns.load_dataset("tips")

# Create a figure with subplots
fig, axs = plt.subplots(2, 2, figsize=(12, 10),dpi=150)

# Subplot 1: Distribution of total bills
sns.histplot(data["total_bill"], kde=True, ax=axs[0, 0])
axs[0, 0].set_title("Distribution of Total Bill")

# Subplot 2: Distribution of tips
sns.histplot(data["tip"], kde=True, ax=axs[0, 1])
axs[0, 1].set_title("Distribution of Tips")

# Subplot 3: Total bill by day
sns.boxplot(x="day", y="total_bill", data=data, ax=axs[1, 0])
axs[1, 0].set_title("Total Bill by Day")

# Subplot 4: Tip by day
sns.boxplot(x="day", y="tip", data=data, ax=axs[1, 1])
axs[1, 1].set_title("Tip by Day")

# Adjust layout
plt.tight_layout()
plt.show()
```
