## Importing Required Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import geom
```

## Visualizing the Probability Mass Function (PMF)


```python
# Set probability of success
p = 0.3

# Define the number of trials for the x-axis
trials = np.arange(1, 21)

# Calculate the PMF for each trial
pmf_values = geom.pmf(trials, p)

# Plot the PMF
plt.figure(figsize=(10, 6),dpi=120)
plt.stem(trials, pmf_values)
plt.title('Geometric Distribution PMF (p=0.3)')
plt.xlabel('Number of Trials (k)')
plt.ylabel('P(X = k)')
plt.show()
```

## Generating Geometric-Distributed Random Numbers

```python
# Set the probability of success
p = 0.3

# Generate random samples
samples = geom.rvs(p, size=1000)

# Plot histogram of samples
plt.figure(figsize=(10, 6),dpi=120)
plt.hist(samples, bins=range(1, max(samples) + 1), density=True, alpha=0.6, color='blue', align='left')
plt.title(f'Histogram of Geometric-distributed samples (p={p})')
plt.xlabel('Number of Trials to First Success')
plt.ylabel('Frequency')
plt.show()

# Sample mean and variance
sample_mean = np.mean(samples)
sample_variance = np.var(samples)
print(f'Sample Mean: {sample_mean:.3f}, Sample Variance: {sample_variance:.3f}')
```

## Example: Modeling Customer Retention

```python
# Probability of purchase on any given visit
purchase_probability = 0.1

# Simulate customer visits until the first purchase
num_customers = 1000
visits_until_purchase = geom.rvs(purchase_probability, size=num_customers)

# Plot the distribution of visits until purchase
plt.figure(figsize=(10, 6),dpi=120)
plt.hist(visits_until_purchase, bins=range(1, max(visits_until_purchase) + 1), density=True, color='purple', alpha=0.6, align='left')
plt.title('Distribution of Visits Until Purchase (p=0.1)')
plt.xlabel('Number of Visits')
plt.ylabel('Frequency')
plt.show()

# Calculate mean visits until purchase
avg_visits = np.mean(visits_until_purchase)
print(f'Average visits until purchase: {avg_visits}')

```
