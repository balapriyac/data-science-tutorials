```python
import math

def poisson_probability(k, lambd):
    return (lambd ** k) * math.exp(-lambd) / math.factorial(k)
```

```python
k = 3
lambd = 2
prob = poisson_probability(k, lambd)
print(f"Probability of exactly {k} events: {prob:.4f}")
```

```python
from scipy.stats import poisson

lambd = 4  # average rate of events
k_values = list(range(10))  # possible event counts (0 to 9)

# PMF and CDF
pmf_values = [poisson.pmf(k, lambd) for k in k_values]
cdf_values = [poisson.cdf(k, lambd) for k in k_values]

print("PMF values:", pmf_values)
print("CDF values:", cdf_values)
```

```python
import matplotlib.pyplot as plt
plt.figure(dpi=180)
plt.bar(k_values, pmf_values, color='skyblue', alpha=0.7, label="PMF")
plt.step(k_values, cdf_values, where='mid', color='orange', label="CDF")
plt.title("Poisson Distribution (λ = 4)")
plt.xlabel("Number of Events (k)")
plt.ylabel("Probability")
plt.legend()
plt.show()
```

```python
# Simulating website visits per minute with λ = 10
data = poisson.rvs(mu=10, size=1000)

# Visualizing the simulation
plt.figure(dpi=180)
plt.hist(data, bins=15, density=True, alpha=0.7, color='skyblue')
plt.title("Simulated Website Traffic (λ = 10)")
plt.xlabel("Number of Visitors")
plt.ylabel("Frequency")
plt.show()
```

```python
prob_5_calls = poisson.pmf(5, 3)
print(f"Probability of exactly 5 calls: {prob_5_calls:.4f}")
```

```python
# Simulating daily system failures (λ = 0.5 failures/day)
failures = poisson.rvs(mu=0.5, size=365)

# Plotting the simulation
plt.figure(dpi=180)
plt.hist(failures, bins=range(5), density=True, color='salmon', alpha=0.7)
plt.title("Simulated Daily System Failures (λ = 0.5)")
plt.xlabel("Number of Failures")
plt.ylabel("Frequency")
plt.show()
```
