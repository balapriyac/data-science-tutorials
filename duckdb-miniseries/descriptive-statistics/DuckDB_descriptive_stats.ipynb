```python
import duckdb
```

```python
duckdb.query("SELECT * FROM read_csv_auto('/content/cab_ride_data.csv') LIMIT 5").df()
```

```python
duckdb.query("""
    SELECT
        AVG(fare) AS avg_fare,
        MEDIAN(fare) AS median_fare,
        AVG(tip) AS avg_tip,
        MEDIAN(tip) AS median_tip,
        AVG(total_distance_km) AS avg_distance,
        MEDIAN(total_distance_km) AS median_distance,
        AVG(duration_min) AS avg_duration,
        MEDIAN(duration_min) AS median_duration
    FROM read_csv('/content/cab_ride_data.csv')
""").df()
```

```python
duckdb.query("""
    SELECT
        STDDEV(fare) AS std_fare,
        VARIANCE(fare) AS var_fare,
        STDDEV(tip) AS std_tip,
        VARIANCE(tip) AS var_tip,
        STDDEV(total_distance_km) AS std_distance,
        VARIANCE(total_distance_km) AS var_distance,
        STDDEV(duration_min) AS std_duration,
        VARIANCE(duration_min) AS var_duration
    FROM read_csv('/content/cab_ride_data.csv')
""").df()
```

```python
duckdb.query("""
    SELECT
        MIN(fare) AS min_fare, MAX(fare) AS max_fare, MAX(fare) - MIN(fare) AS range_fare,
        MIN(tip) AS min_tip, MAX(tip) AS max_tip, MAX(tip) - MIN(tip) AS range_tip,
        MIN(total_distance_km) AS min_distance, MAX(total_distance_km) AS max_distance, MAX(total_distance_km) - MIN(total_distance_km) AS range_distance,
        MIN(duration_min) AS min_duration, MAX(duration_min) AS max_duration, MAX(duration_min) - MIN(duration_min) AS range_duration
    FROM read_csv('/content/cab_ride_data.csv')
""").df()
```

```python
duckdb.query("""
    SELECT
        QUANTILE_CONT(fare, 0.25) AS q1_fare,
        QUANTILE_CONT(fare, 0.50) AS median_fare,
        QUANTILE_CONT(fare, 0.75) AS q3_fare,
        QUANTILE_CONT(total_distance_km, 0.25) AS q1_distance,
        QUANTILE_CONT(total_distance_km, 0.50) AS median_distance,
        QUANTILE_CONT(total_distance_km, 0.75) AS q3_distance
    FROM read_csv('/content/cab_ride_data.csv')
""").df()
```
