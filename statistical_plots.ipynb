```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load the 'tips' dataset from Seaborn
tips = sns.load_dataset("tips")

print(tips.head())
```

```python
plt.figure(dpi=180)
sns.histplot(tips['total_bill'], kde=True, color='blue', bins=20)
plt.title('Distribution of Total Bill')
plt.xlabel('Total Bill')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(dpi=180)
sns.scatterplot(x='total_bill', y='tip', data=tips, hue='size', palette='viridis', size='size', sizes=(20, 200))
plt.title('Scatter Plot of Total Bill vs. Tip')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()
```

```python
plt.figure(dpi=180)
sns.boxplot(x='day', y='total_bill', data=tips, palette='coolwarm')
plt.title('Box Plot of Total Bill by Day')
plt.xlabel('Day of the Week')
plt.ylabel('Total Bill')
plt.show()
```

```python
plt.figure(figsize=(10,8),dpi=180)
sns.pairplot(tips, hue='sex', palette='husl', diag_kind='kde')
plt.show()
```

```python
plt.figure(dpi=180)
sns.violinplot(x='time', y='total_bill', data=tips, palette='muted')
plt.title('Violin Plot of Total Bill by Time of Day')
plt.xlabel('Time of Day')
plt.ylabel('Total Bill')
plt.show()
```

```python
plt.figure(dpi=180)
# Select only numeric features
numeric_features = tips.select_dtypes(include=['number'])

# Compute the correlation matrix
corr_matrix = numeric_features.corr()

# Plot the heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```
